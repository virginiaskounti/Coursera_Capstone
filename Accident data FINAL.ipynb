```python
msk = np.random.rand(len(accident_data)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
train_x = np.asanyarray(train[['Weather_Conditions']])
train_y = np.asanyarray(train[['Severity']])

test_x = np.asanyarray(test[['Weather_Conditions']])
test_y = np.asanyarray(test[['Severity']])


poly = PolynomialFeatures(degree=2)
train_x_poly = poly.fit_transform(train_x)
train_x_poly
```

```python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poly, train_y)

print ('Coefficients: ', clf.coef_)
print ('Intercept: ',clf.intercept_)
```

```python
plt.scatter(train.Weather_Conditions, train.Severity,  color='blue')
XX = np.arange(0.0, 10.0, 0.1)
yy = clf.intercept_[0]+ clf.coef_[0][1]*XX+ clf.coef_[0][2]*np.power(XX, 2)
plt.plot(XX, yy, '-r' )
plt.xlabel("Weather")
plt.ylabel("Severity")
```
